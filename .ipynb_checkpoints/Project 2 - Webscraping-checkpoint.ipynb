```python
from bs4 import BeautifulSoup
import requests
import datetime
```

```python
url = 'https://www.kiva.org/lend/1508099'
response = requests.get(url, headers={'User-Agent': 'Chrome/65.0'})
```

```python
response.status_code
```

```python
print(response.text)
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup)
```

```python
print (soup.prettify())
```

```python
soup.div.prettify()
```

```python
for div in soup.find_all('div'):
    print(div.prettify())
```
